```python
import numpy as np
import matplotlib.pyplot as plt
```

Serial Plot

```python
N_serial = [128, 256, 512, 1024, 2048, 4096, 8192, 16384]
time_serial = [0.06, 0.16, 0.57, 2.21, 8.76, 34.90, 139.60, 558.54]


plt.loglog(N_serial, time_serial, c = 'blue', ls = 'dashed')
plt.xlabel('N')
plt.ylabel('time (s)')
plt.title('Scaling for Serial')
plt.grid()
plt.show()
```

OPEN MP

```python
threads_MP = [1, 2, 4, 8, 16, 32, 64, 128]
time_MP = np.array([166.61, 82.77, 41.71, 21.02, 10.71, 5.60, 3.40, 3.05])
N = 10000


plt.loglog(threads_MP, time_MP/N**2, c = 'red', ls = 'dashed', label = 'OPENMP')
plt.xlabel('# threads')
plt.ylabel(r'Runtime / $N^2$')
plt.title('Scaling for OPENMP, N = 10,000')
plt.grid()
plt.show()

```

MPI

```python
rank_mpi = [1, 2, 4, 8, 16, 32, 64, 128]
time_mpi = np.array([112.65, 58.30, 30.83, 17.11, 10.53, 8.96, 10.20, 16.50])



plt.loglog(rank_mpi, time_mpi/N**2, c = 'green', ls = 'dashed', label = 'MPI')
plt.xlabel('Rank #')
plt.ylabel(r'Runtime / $N^2$')
plt.title('Scaling for MPI, N = 10,000')
plt.grid()
plt.show()
```

```python
#combined

plt.loglog(threads_MP, time_MP/N**2, c = 'red', ls = 'dashed', label = 'OPENMP')
plt.loglog(rank_mpi, time_mpi/N**2, c = 'green', ls = 'dashed', label = 'MPI')
plt.xlabel('N (rank or threads)')
plt.ylabel(r'Runtime / $N^2$')
plt.title('Scaling for OPENMP and MPI, N = 10,000')
plt.grid()
plt.legend()
plt.show()
```

Shared Memory MPI

```python
mpi_rank = [1, 2, 4, 8, 16, 32, 64, 128]
time_shared = np.array([82.73, 47.65, 29.41, 18.97, 9.26, 7.06, 5.79, 10.45])

plt.loglog(rank_mpi, time_mpi/N**2, c = 'green', ls = 'dashed', label = 'MPI')
plt.loglog(mpi_rank, time_shared/N**2, c = 'purple', ls = 'dashed', label = 'Shared MPI')
plt.xlabel('Rank #')
plt.ylabel(r'Runtime / $N^2$')
plt.title('Scaling for MPI, N = 10,000')
plt.grid()
plt.legend()
plt.show()
```
