```python
import numpy as np
import matplotlib.pyplot as plt
```

Serial Plot

```python
N = [128, 256, 512, 1024, 2048, 4096, 8192]
time_serial = [0.129, 0.319, 1.016, 4.440, 16.424, 60.209, 239.316]


plt.loglog(N, time_serial, c = 'blue', ls = 'dashed')
plt.xlabel('N')
plt.ylabel('time (s)')
plt.grid()
plt.show()
```
