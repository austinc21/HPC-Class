```python
n = [10**3, 10**4, 10**5, 10**6, 10**7, 10**8, 10**9]
err = [0.014392654, 0.003272654, 0.000476654, 0.000277054, 0.000200174, 0.000107122, 6.74332*10**-5]
std = [0.07011863, 0.011628012, 0.006202235, 0.00183423, 0.00044909, 0.000153349, 1.60515*10**-5]

plt.plot(n, err, color = 'blue', label = r'$|\bar{\pi}_{estimate} - \pi|$', linestyle = 'dashed')
plt.plot(n, std, color = 'red', label = r'$\sigma$', linestyle = 'dashed')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('n')
plt.title('Error and Standard Deviation Overplotted')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress, t

n = [10**3, 10**4, 10**5, 10**6, 10**7, 10**8, 10**9]
std = [0.07011863, 0.011628012, 0.006202235, 0.00183423, 0.00044909, 0.000153349, 1.60515*10**-5]
logn = np.log10(n)
logstd = np.log10(std)

res = linregress(logn, logstd)
alpha = -res.slope
A = 10**res.intercept

# Got this confidence interval part from ChatGPT, not sure if it's right
tcrit = t.ppf(0.975, df=len(n)-2)
alpha_ci = tcrit*res.stderr

print('Slope = ', res.slope)
print('Intercept = ', res.intercept)
print('alpha = ', alpha, r'+/-', alpha_ci)
print('A = ', A)

# Plot
plt.loglog(n, std, 'o', label='σ data', color = 'red')
plt.loglog(n, A * n**(-alpha), label='Power-law fit', color = 'blue', linestyle = 'dashed')
plt.xlabel('n')
plt.ylabel('σ')
plt.legend()
plt.show()
```
